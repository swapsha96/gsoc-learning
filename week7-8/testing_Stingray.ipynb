```python
def find_all_peaks(dps, model, starting_pars=None, max_post=False,
                      priors=None, fitmethod="L-BFGS-B"):
    """"""
    if not (isinstance(starting_pars, np.ndarray) or isinstance(starting_pars, list)):
        starting_pars = model.parameters
        
    x_0_all = np.array([])
    
    for power in dps.dyn_ps.T:
        ps = Powerspectrum()
        ps.freq = dps.freq
        ps.power = power
        ps.m = dps.m
        ps.df = dps.df
        ps.norm = dps.norm

        _, res = fit_powerspectrum(ps, model, starting_pars, max_post, priors, fitmethod)
        
        x_0 = res.model[0].x_0.value
        
        x_0_all = np.append(x_0_all, x_0)

    return x_0_all
```

```python
dps_fitting_model = models.Lorentz1D(x_0=qpo_init, fwhm=ref_aps.df) + models.PowerLaw1D(amplitude=0.1, x_0=1.5, alpha=0.0) + models.Const1D()
all_peaks = find_all_peaks(dps, dps_fitting_model, fitmethod="CG")
avg_peak = np.mean(all_peaks)
print("Average QPO freq:", avg_peak)
# print(all_peaks)
```

```python
def shift_peaks(dps, all_peaks, new_peak):
    freq = dps.freq
    dps_new = copy.deepcopy(dps)
    n = len(dps.freq)
    m = dps.m
    _, new_idx = find_nearest(dps.freq, new_peak)

    for i in range(m):
        _, old_idx = find_nearest(freq, all_peaks[i])
        diff = new_idx-old_idx
        if diff == 0:
            continue
        elif diff > 0:
            pad_width = (diff, 0)
            start = 0
            end = -1*diff
        else:
            diff *= -1
            pad_width = (0, diff)
            start = diff
            end = n+diff
        dps_new.dyn_ps.T[i] = np.pad(dps_new.dyn_ps.T[i], pad_width, 'constant')[start:end]
#         print(diff, start, end)
    return dps_new
```

```python
dps_new = shift_peaks(dps, all_peaks, avg_peak)
# dps_new_1 = shift_peaks(dps_new, all_peaks, avg_peak)
```

```python
n = len(dps.freq)
range_start, range_stop = int(freq_start*n/max(dps.freq)), int(freq_stop*n/max(dps.freq))

plt.figure(figsize=(8, 6))
extent = min(dps.time), max(dps.time), freq_start, freq_stop
plt.imshow(dps.dyn_ps[:][range_start:range_stop], aspect="auto", origin="lower", vmax=0.001,
           interpolation="none", extent=extent)
plt.title('Old Dynamic Powerspectrum')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Power')

plt.figure(figsize=(8, 6))
extent = min(dps_new.time), max(dps_new.time), freq_start, freq_stop
plt.imshow(dps_new.dyn_ps[:][range_start:range_stop], aspect="auto", origin="lower", vmax=0.001,
           interpolation="none", extent=extent)
plt.title('New Dynamic Powerspectrum')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Power')
```
